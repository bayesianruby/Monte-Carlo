```python

def primes_from_2_to(n):
    """Prime number from 2 to n.
    :param int n: sup bound with ``n >= 6``.
    :return: primes in 2 <= p < n.
    :rtype: list
    """
    sieve = np.ones(n // 3 + (n % 6 == 2), dtype=np.bool)
    for i in range(1, int(n ** 0.5) // 3 + 1):
        if sieve[i]:
            k = 3 * i + 1 | 1
            sieve[k * k // 3::2 * k] = False
            sieve[k * (k - 2 * (i & 1) + 4) // 3::2 * k] = False
    return np.r_[2, 3, ((3 * np.nonzero(sieve)[0][1:] + 1) | 1)]


def van_der_corput(n_sample, base=2):
    """Van der Corput sequence.
    :param int n_sample: number of element of the sequence.
    :param int base: base of the sequence.
    :return: sequence of Van der Corput.
    :rtype: list (n_samples,)
    """
    sequence = []
    for i in range(n_sample):
        n_th_number, denom = 0., 1.
        while i > 0:
            i, remainder = divmod(i, base)
            denom *= base
            n_th_number += remainder / denom
        sequence.append(n_th_number)

    return sequence


def halton(dim, n_sample):
    """Halton sequence.
    :param int dim: dimension
    :param int n_sample: number of samples.
    :return: sequence of Halton.
    :rtype: array_like (n_samples, n_features)
    """
    big_number = 10
    while 'Not enought primes':
        base = primes_from_2_to(big_number)[:dim]
        if len(base) == dim:
            break
        big_number += 1000

    # Generate a sample using a Van der Corput sequence per dimension.
    sample = [van_der_corput(n_sample + 1, dim) for dim in base]
    sample = np.stack(sample, axis=-1)[1:]

    return sample
```

```python
#M random shift for the d dimensions
def generate_shift(M,d):
    """Random shift
    :param int M: number of shifts
    :param int dim: dimension
    :return: matrix of shifts
    :rtype: array_like (M,d)
    """
    list_of_shift = np.zeros((M,d))
    for i in range(M):
        list_of_shift[i] = np.random.rand(d)
    return(list_of_shift)
```

```python

def transfo(mat,d,shift):
    """Shifts the matrix
    :param arr mat: initial matrix
    :param int d: dimension
    :return: matrix shifted
    :rtype: array_like (M,d)
    """
    mat_shifted = (shift+mat)%1
    return(mat_shifted)
    
```

```python


def to_gaussian(vector): #works when the dimension is d=5 
    """converts uniform to gaussian vector
    :param arr vector: uniform vector
    :return: gaussian vector
    :rtype: array_like (d)
    """
    gaussian_vect = np.zeros(5)
    
    gaussian_vect[0]= np.sqrt(-2*np.log(vector[0]))*np.cos(2*np.pi*vector[1])
    gaussian_vect[1]= np.sqrt(-2*np.log(vector[0]))*np.sin(2*np.pi*vector[1])
    
    gaussian_vect[2]= np.sqrt(-2*np.log(vector[2]))*np.cos(2*np.pi*vector[3])
    gaussian_vect[3]= np.sqrt(-2*np.log(vector[2]))*np.sin(2*np.pi*vector[3])
    
    gaussian_vect[4]= np.sqrt(-2*np.log(vector[4]))*np.cos(2*np.pi*vector[5])
    
    return(gaussian_vect)
    

def generate_gaussian(d,n,shift):
    """generates a gaussian vector
    :param int d: dimension
    :return: gaussian vector
    :rtype: array_like (d)
    """
    mat_gauss = np.zeros((n,d-1))
    
    #mat=generate_matrix_u(a,d,n)
    mat= halton(d, n)
    mat_shifted = transfo(mat,d,shift)
    
    for k in range(n):
        mat_gauss[k] = to_gaussian(mat_shifted[k])

    return(mat_gauss)
    
    
```

```python
def estimator(d,n,M,f):
    """Estimates the expectation and the variance of Halton
    :param int d: dimension
    :param int n: number of samples
    :return: expectation, variance
    :rtype: float, float
    """
    I= 0
    list_of_shift = generate_shift(M,d)
    outputsample = np.zeros((M,n))
    var1=0
    s=0
    for m in range(M):
        shift = list_of_shift[m]
        gauss = generate_gaussian(d,n,shift)
        for k in range(n):
            #I += f(gauss[k])
            outputsample[m][k] = f(gauss[k])
            s+=1
        
    I = np.mean(outputsample)
    var = np.var(outputsample)
    
    
    return(I,var)

def MC_estimator(d,n,f):
    """Estimates the expectation and the variance of Monte Carlo
    :param int d: dimension
    :param int n: number of samples
    :return: expectation, variance
    :rtype: float, float
    """
    I= 0
    gauss = np.random.randn(n,d)
    outputsample = np.zeros(n)
    for k in range(n):
        outputsample[k] = f(gauss[k])
    I = np.mean(outputsample)
    var= np.var(outputsample)
    return(I,var)
```
